```python
def calc():
    print('This calculator can calculate: ')
    types={1:'addition',
           2:'subtraction',
           3:'multiplication',
           4:'division',
           5:'raise a number to a power',
           6:'mean',
           7:'standard deviation for a population',
           8:'standard deviation for a sample',
           9:'median',
           10:'area of a circle',
           11:'area of a rectangle',
           12:'area of a triangle',
           13:'perimeter of a circle',
           14:'perimeter of a rectangle',
           15:'perimeter of a triangle'}
    for i in types:
        print(i,types[i])
    function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
    while function!=0:
        if function==1:
            f,s=input('Enter the addition problem you would like to solve: ').split('+')
            f=float(f)
            s=float(s)
            addition(f,s)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==2:
            f,s=input('Enter the subtraction problem you would like to solve: ').split('-')
            f=float(f)
            s=float(s)
            subtraction(f,s)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==3:
            f,s=input('Enter the multiplication problem you would like to solve: ').split('*')
            f=float(f)
            s=float(s)
            multiplication(f,s)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==4:
            f,s=input('Enter the division problem you would like to solve: ').split('/')
            f=float(f)
            s=float(s)
            division(f,s)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==5:
            print('Notation is f^s')
            f,s=input('Enter the power problem you would like to solve: ').split('^')
            f=float(f)
            s=float(s)
            power(f,s)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==6:
            num=(input('Enter numbers seperated by a comma: ')).split(',')
            num=[float(i) for i in num]
            mean(num)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==7:
            num=(input('Enter numbers seperated by a comma: ')).split(',')
            num=[float(i) for i in num]
            standard_deviationpop(num)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==8:
            num=(input('Enter numbers seperated by a comma: ')).split(',')
            num=[float(i) for i in num]
            standard_deviationsamp(num)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==9:
            num=(input('Enter numbers seperated by a comma: ')).split(',')
            num=[float(i) for i in num]
            median(num)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==10:
            r=float(input('Enter the radius: '))
            area_circle(r)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==11:
            l,w=input('Enter the length and width spereated by a comma: ').split(',')
            l=float(l)
            w=float(w)
            area_rectangle(l,w)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==12:
            a,b,c=input('Enter the three sides of the triangle seperated by a comma: ')
            a=float(a)
            b=float(b)
            c=float(c)
            area_triangle(a,b,c)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==13:
            r=float(input('Enter the radius: '))
            perimeter_circle(r)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==14:
            l,w=input('Enter the length and width spereated by a comma: ').split(',')
            l=float(l)
            w=float(w)
            perimeter_rectangle(l,w)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        elif function==15:
            a,b,c=input('Enter the three sides of the triangle seperated by a comma: ')
            a=float(a)
            b=float(b)
            c=float(c)
            perimeter_triangle(a,b,c)
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
        else:
            print('Invalid Input. Try Again.')
            function=int(input('Type the number of the function you would like to preform. Type 0 to quit: '))
def addition(f,s):
    print(f+s)
def subtraction(f,s):
    print(f-s)
def multiplication(f,s):
    print(f*s)
def division(f,s):
    print(f/s)
def power(f,s):
    print(f**s)
def mean(args):
    print((sum(args)/len(args)))
def standard_deviationpop(args):
    m=(sum(args)/len(args))
    p=0
    for i in  args:
        p=((i-m)**2)+p
    print((p/len(args))**.5)
def standard_deviationsamp(args):
    m=(sum(args)/len(args))
    p=0
    for i in  args:
        p=((i-m)**2)+p
    print((p/(len(args)-1))**.5)
def median(args):
    if len(args)%2==1:
        print(args[int(len(args)/2)])
    else:
        print((args[int(len(args)/2-.5)]+args[int(len(args)/2+.5)])/2)
def area_circle(r):
    import math
    print(math.pi*r**2)
def area_rectangle(l,w):
    print(l*w)
def area_triangle(a,b,c):
    s=(a+b+c)/2
    print((s*(s-a)*(s-b)*(s-c)**.5))
def perimeter_circle(r):
    import math
    print(math.pi*r*2)
def perimeter_rectangle(l,w):
    print(2*(l+w))
def perimeter_triangle(a,b,c):
    print(a+b+c)
```
